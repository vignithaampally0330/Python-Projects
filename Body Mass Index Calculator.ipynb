```python
print()

#opening message
print("Welcome to our Body Mass Index Calculator Website!")
print()

# Stores the user's full name
name = input('Please enter your full name: ')
print()

# Stores the user's age
age = input('Please enter your age: ')
print()

# Stores the user's sex
sex = input('Please enter your sex: ')
print()

print('----------------------------------------------------------------------------------')
print()

# Outputs the message with the user's name
print('Hi ' + name + '! We are happy to assist you!')
print()
print('----------------------------------------------------------------------------------')
print()

# Provides further instructions for the user
print('Please answer the following questions as follows:')
print()

# Stores the weight question output
weight_question = input('Would you like to enter your weight in kgs or lbs? Please enter 1 to choose kgs and 2 to choose lbs.\n')
print()

# Declare the finalized weight variable
finalized_weight = 0
# Declare the finalized height variable
finalized_height = 0

# Conditional Statements to convert weight from kgs to lbs.
if weight_question == '1':
    # Stores the kg weight response 
    kgs_response = input('Please enter your weight in kgs. Please only enter the numerical value.\n')
    # Converts the string to float to do mathematical calculations
    kgs_response = float(kgs_response)
    # Mutlitply kgs value with 2.205 to get weight in lbs and store it in finalized_weight variable.
    finalized_weight = kgs_response * 2.205
elif weight_question == '2':
    # Stores the lbs weight response
    lbs_response = input('Please enter your weight in lbs. Please only enter the numerical value.\n')
    # Converts the string to float to do mathematical calculations
    lbs_response = float(lbs_response)
    # Store the lbs response value in the finalized_weight variable
    finalized_weight = lbs_response
# Error Message if the above statements are not satisfied
else:
    print()
    print('INVALID INPUT')

print()
print('----------------------------------------------------------------------------------')
print()

# Stores the height question output
height_question = input('Would you like to enter your height in inches or feet? Please enter 1 to choose inches and 2 to choose feet.\n')
print()

# Conditional Statements to convert height from feet to inches.
if height_question == '2':
    # Stores the height in feet response
    feet_response = input('Please enter your height below in feet. Please enter in the following form: example: 5 feet 9.5 inches -> 5_9.5 Please make sure to add an underscore between the numbers as shown in numerical form.\n')
    # Splits the response in two values as shown below
    feet_value = feet_response.split('_')[0]
    inch_value = feet_response.split('_')[1]
    # Converts the string to float to peform mathematical calculations
    feet_value = float(feet_value)
    inch_value = float(inch_value)
    # Convert the height in feet to height in inches using the formuala below
    finalized_height = (feet_value * 12) + inch_value
elif height_question == '1':
    # Stores the height in inches response
    inches_response = input('Please enter your height in inches. Please only enter the numerical value.\n')
    # Converts the string to float to perform mathematical calculations
    inches_response = float(inches_response)
    # Store the inches response in the finalized_height variable
    finalized_height = inches_response
# Error Message if the above statements are not satisfied
else:
    print()
    print('INVALID INPUT')   

print()
# print(finalized_weight)
# print(finalized_height)

# Calculate the BMI and round it to 4 decimal places using the formula below
BMI = (finalized_weight * 703) / (finalized_height * finalized_height)
BMI = round(BMI,4)

# Message classifiation based on the BMI as shown below
message = ''
# If BMI is greater than 0
if BMI > 0:
    # If BMI is less than 18.5 and greater than 0
    if (BMI < 18.5):
        message = 'you are underweight.'
    # If BMI is less than or equal to 24.9 assuming the previous conidition not meet
    elif (BMI <= 24.9):
        message = 'you are normal weight.'
    # If BMI is less than or equal to 29.9 assuming the previous conidition not meet
    elif (BMI <= 29.9):
        message = 'you are overweight.'
    # If BMI is less than or equal to 34.9 assuming the previous conidition not meet
    elif (BMI <= 34.9):
        message = 'you are obese.'
    # If BMI is less than or equal to 39.9 assuming the previous conidition not meet
    elif (BMI <= 39.9):
        message = 'you are severely obese.'
    # If BMI is greater than 39.9 assuming the previous conidition not meet
    else:
        message = 'you are morbidly obese.'
# Error Message if the above statements are not satisfied
else:
    print('Enter valid inputs')

# Calculate the body fat percentage using the formula below
body_fat_percentage = (1.20 * BMI) / (0.23 * int(age))

print()
print('----------------------------------------------------------------------------------')
print()
# Displaying the finalized details below
print('YOUR DETAILS: ')
print()
print('Name: ' + name)
print('Age: ' + age)
print('Sex: ' + sex)
print('Height: ' + str(finalized_height) + ' inches.')
print('Weight: ' + str(finalized_weight) + ' lbs.')
print()
print('Body Fat Percentage: ' + str(round(body_fat_percentage,2)) + '%')
print()
print('Your BMI is ' + str(BMI) + '.')
print()
print('Our analysis is that ' + message)
print()
```

```python
def bmi_calculation(user_full_name, user_age, user_sex, height_inches, weight_lbs):

    finalized_weight = weight_lbs
    finalized_height = height_inches

    BMI = (finalized_weight * 703) / (finalized_height * finalized_height)
    
    BMI = round(BMI,4)
    
    message = ''
    if BMI > 0:
        if (BMI < 18.5):
            message = 'you are underweight.'
        elif (BMI <= 24.9):
            message = 'you are normal weight.'
        elif (BMI <= 29.9):
            message = 'you are overweight.'
        elif (BMI <= 34.9):
            message = 'you are obese.'
        elif (BMI <= 39.9):
            message = 'you are severely obese.'
        else:
            message = 'you are morbidly obese.'
    else:
        print('Enter valid inputs')

    body_fat_percentage = (1.20 * BMI) / (0.23 * int(user_age))

    print()
    print('YOUR DETAILS: ')
    print()
    print('Name: ' + user_full_name)
    print('Age: ' + str(user_age))
    print('Sex: ' + user_sex)
    print('Height: ' + str(finalized_height) + ' inches.')
    print('Weight: ' + str(finalized_weight) + ' lbs.')
    print()
    print('Body Fat Percentage: ' + str(round(body_fat_percentage,2)) + '%')
    print()
    print('Your BMI is ' + str(BMI) + '.')
    print()
    print('Our analysis is that ' + message)
    print()
```
